```python
# RE-IMPORTS
import numpy as np
import pandas as pd
import rasterio
from rasterio.transform import xy
import os

# Raster to DataFrame converter
def raster_to_df(raster_path, value_name):
    with rasterio.open(raster_path) as src:
        data = src.read(1)
        transform = src.transform
        nodata = src.nodata

        rows, cols = np.where(data != nodata)
        lats, lons = [], []
        values = data[rows, cols]

        for r, c in zip(rows, cols):
            lon, lat = xy(transform, r, c)
            lats.append(lat)
            lons.append(lon)

        return pd.DataFrame({
            'Latitude': lats,
            'Longitude': lons,
            value_name: values
        })

# Merge all raster .asc into one master_df
file_map = {
    "fsoildep0_25.asc": "SoilDepth_0_25",
    "fsoildep25_50.asc": "SoilDepth_25_50",
    "fsoildep50_75.asc": "SoilDepth_50_75",
    "fsoildep75_100.asc": "SoilDepth_75_100",
    "fsoildep100_150.asc": "SoilDepth_100_150",
    "fsoildep150_200.asc": "SoilDepth_150_200",
    "meantocd.asc": "OrganicCarbonDensity",
    "meanticd.asc": "InorganicCarbonDensity",
    "fsandy.asc": "Texture_Sandy",
    "floamy.asc": "Texture_Loamy",
    "fclayey.asc": "Texture_Clayey",
    "fclayskeletal.asc": "Texture_Clayskeletal"
}

base_path = "/content/extracted_files"
master_df = None

for fname, col_name in file_map.items():
    fpath = os.path.join(base_path, fname)
    df = raster_to_df(fpath, col_name)

    if master_df is None:
        master_df = df
    else:
        master_df = pd.merge(master_df, df, on=["Latitude", "Longitude"], how="outer")

print("✅ master_df shape:", master_df.shape)
master_df.head()
```

```python
# Assigning  soil texture from binary mask columns
def assign_texture(row):
    if row.get("Texture_Sandy", 0) == 1:
        return "sandy"
    elif row.get("Texture_Loamy", 0) == 1:
        return "loamy"
    elif row.get("Texture_Clayey", 0) == 1:
        return "clayey"
    elif row.get("Texture_Clayskeletal", 0) == 1:
        return "clayskeletal"
    return "Unknown"

master_df["SoilTexture"] = master_df.apply(assign_texture, axis=1)

# Calculate average depth from available soil layers
depth_cols = [col for col in master_df.columns if "SoilDepth_" in col]
master_df["AvgDepth"] = master_df[depth_cols].mean(axis=1)
```

```python
# Filter valid rows for machine learning
usable_df = master_df[
    (master_df["SoilTexture"] != "Unknown") &
    (master_df["OrganicCarbonDensity"].notnull()) &
    (master_df["AvgDepth"].notnull())
].copy()

print("✅ Usable ML rows:", usable_df.shape[0])
```

```python
# Apply fertilizer rule-based logic
usable_df[["CarbonClass", "Urea_Rec", "DAP_Rec", "MOP_Rec"]] = usable_df.apply(recommend_fertilizer, axis=1)
```

```python
def recommend_fertilizer(row):
    texture = row["SoilTexture"]
    carbon = row["OrganicCarbonDensity"]

    if pd.isna(carbon):
        carbon_class = "Unknown"
    elif carbon < 1.0:
        carbon_class = "Low"
    else:
        carbon_class = "High"

    if texture == "sandy":
        return pd.Series([carbon_class, "High", "Medium", "High"])
    elif texture == "loamy":
        return pd.Series([carbon_class, "Medium", "Medium", "Medium"])
    elif texture == "clayey":
        return pd.Series([carbon_class, "Low", "High", "Low"])
    elif texture == "clayskeletal":
        return pd.Series([carbon_class, "Medium", "High", "Medium"])

    return pd.Series([carbon_class, "Unknown", "Unknown", "Unknown"])

usable_df[["CarbonClass", "Urea_Rec", "DAP_Rec", "MOP_Rec"]] = usable_df.apply(recommend_fertilizer, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode SoilTexture
enc_texture = LabelEncoder()
usable_df["TextureEnc"] = enc_texture.fit_transform(usable_df["SoilTexture"])

# Encode fertilizer labels
enc_urea = LabelEncoder()
enc_dap  = LabelEncoder()
enc_mop  = LabelEncoder()

usable_df["UreaEnc"] = enc_urea.fit_transform(usable_df["Urea_Rec"])
usable_df["DAPEnc"]  = enc_dap.fit_transform(usable_df["DAP_Rec"])
usable_df["MOPEnc"]  = enc_mop.fit_transform(usable_df["MOP_Rec"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Features and targets
X = usable_df[["TextureEnc", "OrganicCarbonDensity", "AvgDepth"]].values
y_urea = usable_df["UreaEnc"].values
y_dap  = usable_df["DAPEnc"].values
y_mop  = usable_df["MOPEnc"].values

# Split data
X_train, X_test, y_train_urea, y_test_urea = train_test_split(X, y_urea, test_size=0.2, random_state=42)
_, _, y_train_dap, y_test_dap = train_test_split(X, y_dap, test_size=0.2, random_state=42)
_, _, y_train_mop, y_test_mop = train_test_split(X, y_mop, test_size=0.2, random_state=42)

# Train RandomForest models
rf_urea = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train_urea)
rf_dap  = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train_dap)
rf_mop  = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train_mop)

```

```python
# Predict
y_pred_urea = rf_urea.predict(X_test)
y_pred_dap  = rf_dap.predict(X_test)
y_pred_mop  = rf_mop.predict(X_test)

# Classification reports
print("📊 UREA CLASSIFICATION:")
print(classification_report(y_test_urea, y_pred_urea, target_names=enc_urea.classes_))

print("📊 DAP CLASSIFICATION:")
print(classification_report(y_test_dap, y_pred_dap, target_names=enc_dap.classes_))

print("📊 MOP CLASSIFICATION:")
print(classification_report(y_test_mop, y_pred_mop, target_names=enc_mop.classes_))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_importance(model, feature_names, title):
    importances = model.feature_importances_
    indices = np.argsort(importances)

    plt.figure(figsize=(5, 3))
    plt.barh(range(len(indices)), importances[indices])
    plt.yticks(range(len(indices)), [feature_names[i] for i in indices])
    plt.title(title)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

features = ["TextureEnc", "OrganicCarbonDensity", "AvgDepth"]

plot_importance(rf_urea, features, "Feature Importance – UREA")
plot_importance(rf_dap, features, "Feature Importance – DAP")
plot_importance(rf_mop, features, "Feature Importance – MOP")
```

```python
import xgboost as xgb

# Use previously encoded features
X = usable_df[["TextureEnc", "OrganicCarbonDensity", "AvgDepth"]].values
y_urea = usable_df["UreaEnc"].values
y_dap  = usable_df["DAPEnc"].values
y_mop  = usable_df["MOPEnc"].values

# Split data
X_train, X_test, y_train_urea, y_test_urea = train_test_split(X, y_urea, test_size=0.2, random_state=42)
_, _, y_train_dap, y_test_dap = train_test_split(X, y_dap, test_size=0.2, random_state=42)
_, _, y_train_mop, y_test_mop = train_test_split(X, y_mop, test_size=0.2, random_state=42)

# Train XGBoost models
xgb_urea = xgb.XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='mlogloss')
xgb_dap  = xgb.XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='mlogloss')
xgb_mop  = xgb.XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='mlogloss')

xgb_urea.fit(X_train, y_train_urea)
xgb_dap.fit(X_train, y_train_dap)
xgb_mop.fit(X_train, y_train_mop)
```

```python
from sklearn.metrics import classification_report

# Predict
y_pred_urea = xgb_urea.predict(X_test)
y_pred_dap  = xgb_dap.predict(X_test)
y_pred_mop  = xgb_mop.predict(X_test)

# Reports
print("📊 XGBoost – UREA")
print(classification_report(y_test_urea, y_pred_urea, target_names=enc_urea.classes_))

print("📊 XGBoost – DAP")
print(classification_report(y_test_dap, y_pred_dap, target_names=enc_dap.classes_))

print("📊 XGBoost – MOP")
print(classification_report(y_test_mop, y_pred_mop, target_names=enc_mop.classes_))
```

```python
import matplotlib.pyplot as plt

def plot_xgb_importance(model, title):
    xgb.plot_importance(model)
    plt.title(title)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_xgb_importance(xgb_urea, "XGBoost Feature Importance – UREA")
plot_xgb_importance(xgb_dap, "XGBoost Feature Importance – DAP")
plot_xgb_importance(xgb_mop, "XGBoost Feature Importance – MOP")
```

```python
import gradio as gr
import numpy as np

# Class labeling
texture_labels = list(enc_texture.classes_)

def predict_fertilizer(texture, carbon, depth):
    try:
        tex_encoded = enc_texture.transform([texture])[0]
    except:
        return "Invalid Texture", "Invalid", "Invalid"

    X_input = np.array([[tex_encoded, carbon, depth]])

    urea_pred = enc_urea.inverse_transform(xgb_urea.predict(X_input))[0]
    dap_pred  = enc_dap.inverse_transform(xgb_dap.predict(X_input))[0]
    mop_pred  = enc_mop.inverse_transform(xgb_mop.predict(X_input))[0]

    return urea_pred, dap_pred, mop_pred

# Gradio interfacing
demo = gr.Interface(
    fn=predict_fertilizer,
    inputs=[
        gr.Dropdown(choices=texture_labels, label="Soil Texture"),
        gr.Slider(minimum=0.0, maximum=10.0, step=0.01, value=1.25, label="Organic Carbon Density"),
        gr.Slider(minimum=0, maximum=2000, step=10, value=300, label="Average Depth")
    ],
    outputs=[
        gr.Textbox(label="Urea Recommendation"),
        gr.Textbox(label="DAP Recommendation"),
        gr.Textbox(label="MOP Recommendation")
    ],
    title="🌱 Fertilizer Recommendation System",
    description="Enter soil conditions to predict optimal fertilizer quantities."
)

demo.launch(share=True)

```
